```python
import plotly.express as px
```

```python
df = px.data.gapminder()
df.head()
```

```python
df = df.query('year == 2007')
df.head()
```

```python
px.histogram(df, x='lifeExp', y='pop', marginal='rug', hover_name='country', height=700, color='continent'
            )
```

```python
px.histogram(df, x='continent', y='pop', hover_name='country', marginal='rug', height=700, color='continent')
```

```python
px.bar(df, x='continent', y='pop', hover_name='country', height=700, color='lifeExp')
```

```python
px.sunburst(df, path=['continent', 'country'], hover_name='country', height=700, color='lifeExp')
```

```python
px.treemap(df, path=['continent', 'country'], hover_name='country', height=700, color='lifeExp')
```

```python
px.choropleth(df, locations='iso_alpha', hover_name='country', hover_data=df.columns, height=700, color='lifeExp')
```

```python
px.choropleth(df, locations='iso_alpha', hover_name='country', hover_data=df.columns, height=700, color='lifeExp', projection='natural earth')
```
